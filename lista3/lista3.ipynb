```python
from chefboost import Chefboost as chef
```

```python
import pandas as pd

# importando e imprimindo a database
df = pd.read_csv("sample_data/credit.csv", sep=';')
df
```

```python
# c4.5
config = {'algorithm': 'C4.5'}
model_c45 = chef.fit(df, config = config, target_label = 'Risco')
```

```python
# testando a previsão da árvore
prediction = chef.predict(model_c45, param = ['Boa', 'Alta', 'Nenhuma', '<15000'])
prediction
```

```python
# id3
config = {'algorithm': 'ID3'}
model_id3 = chef.fit(df, config = config, target_label = 'Risco')
```
