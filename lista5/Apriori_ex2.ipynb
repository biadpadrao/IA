```python
import pandas as pd
from apyori import apriori
```

```python
base = pd.read_csv('MercadoSim.csv', ';', encoding='cp1252', header=None)
base
```

```python
base.shape
```

```python
transacoes = []
for i in range(len(base)):
  # print(i)
  # print(base.values[i, 0])
  transacoes.append([str(base.values[i, j]) for j in range(base.shape[1])])
```

```python
transacoes
```

```python
type(transacoes)
```

```python
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.8)
saida = list(regras)
print(len(saida))
print(saida)
```

```python
print(saida[2])
```

```python
Antecedente = []
Consequente = []
suporte = []
confianca = []
lift = []

for resultado in saida:
  s = resultado[1]
  result_rules = resultado[2]
  for result_rule in result_rules:
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    if 'nan' in a or 'nan' in b: continue
    if len(a) == 0 or len(b) == 0: continue
    Antecedente.append(a)
    Consequente.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
    RegrasFinais = pd.DataFrame({'Antecedente': Antecedente, 'Consequente': Consequente, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
```
